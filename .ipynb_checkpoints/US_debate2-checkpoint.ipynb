# split data into 3 groups 

```python
index_list = []
for i in range(len(text_list)):
    if text_list[i] in ("CLINTON",'TRUMP','HOLT'):
        index_list.append(i)
print(len(index_list))
```

```python
from collections import defaultdict
result_dict = defaultdict(list)
for index in range(len(index_list)-1):
    value = text_list[index_list[index]+1:index_list[index+1]]
    result_dict[text_list[index_list[index]]].append(value)
print(len(result_dict))
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

## text input

```python
# pure lowercase text without punctuations
def get_text_input(person):
    temp = [item.lower() for sublist in result_dict[person] for item in sublist]
    text = ' '.join(temp)
    return text 
```

```python
clinton_text = get_text_input('CLINTON')
trump_text = get_text_input('TRUMP')
```

## list input

```python
# one string as one item in list
def get_list_input(person):
    lst = [' '.join(s) for s in result_dict[person]]
    return lst
```

```python
clinton_list = get_list_input('CLINTON')
trump_list = get_list_input('TRUMP')
```

# countvectorizer + TFIDFvectorizer 

```python
def generate_cvec_c_vocab(text): #text should be a list of strings
    cvec = CountVectorizer(stop_words='english',ngram_range=(1,1))
    c = cvec.fit_transform(text)
    vocab_c = cvec.vocabulary_
    reversed_vocab_c = dict((v,k) for (k,v) in vocab_c.items())
    return cvec, c, reversed_vocab_c
```

```python
def generate_tfidfvec_tfidf_vocab(text):
    tfidfvec = TfidfVectorizer(stop_words='english',ngram_range=(1,1))
    tfidf = tfidfvec.fit_transform(text)
    vocab_t = tfidfvec.vocabulary_
    reversed_vocab_t = dict((v,k) for (k,v) in vocab_t.items())
    return tfidfvec,tfidf, reversed_vocab_t
```

```python
def get_topK_vocab(k, reversed_vocab,vector):
    topwords= set()
    for i in range(vector.shape[0]):
        sublist = vector[i,:].todense()
        sorted_sublist = np.argsort(sublist).tolist()[0]
        topword_index = sorted_sublist[0:k]
        for index in topword_index:
            topwords.add(reversed_vocab[index])
        #print(len(topwords))
    return topwords
```

```python
def convert_set2str(vocab):
    count_string = ''
    for word in vocab:
        count_string+=word +" "
    return count_string
```

# count frequency 

```python
def get_count_words(vector, vec):
    ranked_tuple= list(zip(vec.get_feature_names(),
    np.asarray(vector.sum(axis=0)).ravel()))
    ranked_result= sorted(ranked_tuple, key=lambda x: x[1],reverse=True)
    return ranked_result
```
